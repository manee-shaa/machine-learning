```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
sp=pd.read_csv(r"C:\Users\manis\Downloads\spam.csv")
sp
```

```python
sp.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
sp['Category']=LabelEncoder().fit_transform(sp['Category'])
```

```python
sp
```

```python
y=sp[['Category']]
y
```

```python
x=sp['Message']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
import numpy as np
```

```python
x_train=np.array(x_train).ravel()
```

```python
x_train
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer()
x_train_counter=cv.fit_transform(x_train)
x_train_counter
```

```python
pd.DataFrame(x_train_counter.toarray(),columns=cv.get_feature_names_out())
```

```python
model=MultinomialNB()
model.fit(x_train_counter,y_train)
```

```python
train_pred=model.predict(x_train_counter)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(train_pred,y_train)
```

```python
test_pred=model.predict(cv.transform(x_test))
```

```python
accuracy_score(test_pred,y_test)
```
