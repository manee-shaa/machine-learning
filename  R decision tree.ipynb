```python
import pandas as pd
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
data=pd.read_csv(r"C:\Users\manis\Downloads\Automobile_data.csv")
data
```

```python
data.isna().sum()
```

```python
mean=data.price.mean()
data["price"]=data["price"].fillna(mean)
```

```python
from word2number import w2n
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
```

```python
data
```

```python
data.drop('index',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data['company']=LabelEncoder().fit_transform(data['company'])
data['body-style']=LabelEncoder().fit_transform(data['body-style'])
data['engine-type']=LabelEncoder().fit_transform(data['engine-type'])
```

```python
data
```

```python
y=data[['price']]
```

```python
y
```

```python
x=data.drop('price',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train
```

```python
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
test_pred=model.predict(x_test)
```

```python
train_pred=model.predict(x_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(train_pred,y_train)
```

```python
r2_score(test_pred,y_test)
```

```python
from sklearn import tree
```

```python
model2= DecisionTreeRegressor(max_depth=2,min_samples_leaf=3,min_samples_split=4,criterion='absolute_error')
model2.fit(x_train,y_train)
```

```python
train_pred=model2.predict(x_train)
train_pred
```

```python
test_pred=model2.predict(x_test)
test_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```
